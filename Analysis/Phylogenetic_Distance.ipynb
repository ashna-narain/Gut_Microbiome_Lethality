```python
from Bio import Phylo
import pandas as pd
```

```python
# Load the phylogenetic tree
tree_file = "Gut_Microbiome_phyliptree.phy"  # Replace with the correct path
tree = Phylo.read(tree_file, "newick")

# Get the list of all taxa
taxa = [clade.name for clade in tree.get_terminals() if clade.name]

# Compute the pairwise phylogenetic distance matrix
distance_matrix = pd.DataFrame(index=taxa, columns=taxa)

for taxon1 in taxa:
    for taxon2 in taxa:
        distance_matrix.loc[taxon1, taxon2] = tree.distance(taxon1, taxon2)

# Convert the distance matrix to numeric values
distance_matrix = distance_matrix.astype(float)

# Save or print the matrix
print(distance_matrix)
distance_matrix.to_csv("Gut_Microbiome_Phylogenetic Distances.csv")
```
